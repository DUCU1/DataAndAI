# 3. Create a pie chart of the number of pieces of fruit the students eat per day.

```python
import matplotlib.pyplot as plt
x = studenq.Fruit.value_counts(dropna=False).sort_index()
l = studenq.Fruit.unique()
plt.figure()
plt.pie(x, labels=l)
plt.title('Fruit Distribution')
plt.show()
```

# 4. How divided are the students' views on the number of years Biden will live in the White House? (barplot or hist?)

```python
x = studenq['Biden'].value_counts().sort_index()
l = studenq.Biden.dropna().unique()
plt.figure()
plt.bar(l, x)
plt.title("Biden's Years In White House")
plt.xlabel('Years')
plt.ylabel('Number')
plt.show()
```

# 5. Visualize the students' favourite leisure activities depending on their place of residence (Antwerp or not).

```python
ct= pd.crosstab(studenq['Antwerp Address'], studenq.Leisure,margins=False)
plt.figure()
ct.plot.bar(title='Leisure', stacked=True)
plt.xlabel('Antwerp Address')
plt.show()
```
