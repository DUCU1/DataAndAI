1. How many % of students have 3 or fewer siblings?

```python
def all_freq(x):
    t_abs = x.value_counts(dropna=False).sort_index()
    t_rel = (x.value_counts(dropna=False, normalize=True).sort_index()*100).round(1)
    t_abs_cum = x.value_counts(dropna=False).sort_index().cumsum()
    t_rel_cum = (x.value_counts(dropna=False, normalize=True).sort_index().cumsum()*100).round(1)
    return pd.DataFrame({'abs freq':t_abs,'rel freq':t_rel,'abs cum freq':t_abs_cum,'relcum freq':t_rel_cum})

print(all_freq(studenq.Siblings).iloc[3][3])
```

3. Create a pie chart of the number of pieces of fruit
the students eat per day.

```python
import matplotlib.pyplot as plt
x = studenq.Fruit.value_counts(dropna=False).sort_index()
I = studenq['Fruit'].unique()
plt.figure()
plt.pie(x, labels=I)
plt.title('Fruit Distribution')
plt.show()
```

4. How divided are the students' views on the number
of years Biden will live in the White House? (barplot
or hist?)

```python
x = studenq.Biden.value_counts(dropna=False).sort_index()
I = studenq['Biden'].unique()
plt.figure()
plt.bar(I, x, color='yellow')
plt.title('Biden will be president')
plt.xlabel('Years')
plt.ylabel('Number of Students')
plt.show()
```
