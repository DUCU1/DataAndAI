# Realiability

```python
x = linkedin['# Connections']
y = linkedin.Salary
model = np.polyfit(x,y,1)
print(model)
predict = np.poly1d(model)
xx = np.arange(x.min(), x.max()+1, (x.max()-x.min())/100)
yy = predict(xx)
plt.figure()
plt.scatter(x,y)
plt.plot(xx,yy,color='red')

#
y_pred = predict(x)
se = math.sqrt(((y_pred - y)**2).mean())
plt.fill_between(xx,yy- 2 * se,yy + 2 * se, color ='#FFFF0080')
plt.show()
```
